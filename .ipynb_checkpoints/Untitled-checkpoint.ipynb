```python
import requests
import bs4
import re
```

```python
pri = "https://www.boxofficemojo.com/yearly/chart/?yr=2017&p=.htm"
res = requests.get(pri)
```

```python
soup = bs4.BeautifulSoup(res.text, "html.parser")
type(soup)
```

```python
print (soup.prettify()[0:100])
```

```python
for link in soup.find_all('a'):
    print (link.get('href'))
```

```python
for link in soup.find_all('a', attrs={'href':re.compile("^http")}):
    print (link)
```

```python
file = open('parsed_data.txt', 'wb')
for link in soup.find_all('a', attrs={'href':re.compile("^http")}):
    soup_link = str(link)
    print (soup_link)
    file.write(soup_link)
    file.flush()
    file.close()
```

```python
soup.find_all(['ul', 'b'])
```

```python
l = re.compile('l')
for tag in soup.find_all(l):
    print(tag.name)
```
